```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = '../data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app, encoding='utf-8')
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
print('row number:', len(app_train.index),
      '\ncolumn number:', len(app_train.columns))
```

#### 列出所有欄位

```python
print(app_train.columns, app_train.index)
```

#### 截取部分資料

```python
app_train.loc[20:30, ['SK_ID_CURR', 'TARGET', 'NAME_CONTRACT_TYPE']]
```

```python
# random selet from dataFrame
app_train.loc[np.random.randint(0,len(app_train.index)/2):np.random.randint(len(app_train.index)/2,len(app_train.index)), 
              [app_train.columns[np.random.choice(len(app_train.columns))] for i in range(np.random.choice(len(app_train.columns)//5))]]
```
