```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# k-trees to check unique_data ratio estimation by bagging or boostrap method
data = np.arange(1000)
k = 1000
k_result = [len(np.unique(np.random.choice(data, len(data)))) / len(data) for i in range(k)]
k_mean = np.mean(k_result)
k_std = np.std(k_result)
print('k:{} trees bagging with unique sampling data ratio:\n"{:.3} +/- {:.3} with 95% confidence level"'.format(k, k_mean, 2*k_std))
```

```python
plt.figure(figsize=(5, 5))
plt.pie(x = [k_mean, 1 - k_mean], labels= ['unique_selected_sample', 'rest_not_selected_sample'], 
        explode=[0.1, 0], autopct='%1.2f%%', shadow=True, startangle=90)
plt.show()
```
