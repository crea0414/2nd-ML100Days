## 請搜尋目標為多標籤問題 (Multi-label) 的機器學習案例，了解其資料來源、目標以及評估指標為何
## Hint: 服飾

[Multi-Label-Classification](https://towardsdatascience.com/fast-ai-season-1-episode-3-a-case-of-multi-label-classification-a4a90672a889)

其資料是屬影像類型資料，來源是fastai提供(也可能是衛星收集的地面照片之類)。  
目的在於分類拍攝的單張地面照片其天氣狀態(hase/clear)以及地面分佈型態(water/primary/agriculture)等。  
將一照片分成17個類別，以One Hot Encoding的方式達到multi-label。  
值得一提是分類的部分multi-label 一對多的activation是用sigmoid，對比multi-class一對一的分類activation採用softmax。   

評估指標採用f2  
(f-Beta，In the f-Beta score , Beta says how much you weight false negatives vs false positives ? Here in f-Beta , Beta value is 2)
