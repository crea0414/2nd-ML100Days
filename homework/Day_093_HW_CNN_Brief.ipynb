```python
import keras
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create CNN model

```python
keras.backend.clear_session()
model = keras.models.Sequential()
#first convolution layer and maxpool
model.add(keras.layers.Conv2D(32, (3, 3), input_shape=(28, 28, 1)))
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.MaxPool2D(2, 2))
#second convolution layer and maxpool
model.add(keras.layers.Conv2D(25, (3, 3))) #don't specify input_shape after first layer
model.add(keras.layers.Activation('relu'))
model.add(keras.layers.MaxPool2D(2, 2))
#flattern layer
model.add(keras.layers.Flatten())
#full connected layer
model.add(keras.layers.Dense(units=100))
model.add(keras.layers.Activation('relu'))
#output layer
model.add(keras.layers.Dense(units=10))
model.add(keras.layers.Activation('softmax'))
model.summary()
```

- Flatten Dense Layer vs Con with Flattern Dense Layer: The main difference is Con with Flattern will keep shape information (width / height / channels)
- Pooling Layer has no parameters
